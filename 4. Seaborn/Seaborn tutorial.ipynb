##### plotting a graph to differentiate between smokers and non smokers using hue.

```python
# lmplot is used to plot a corresponding x y graph  and mark the corresponding points on the graph

import seaborn as sb
tips = sb.load_dataset('tips') 
tips.head()
```

<font color='red'>hue - hue will differentiate the plot on the basis of the parameters passed in it.
suppose we give hue = 'smoker' then hue will differentiate the graph on the basis of smokers and non smokers.

```python
import matplotlib.pyplot as plt
import seaborn as sb
tips=sb.load_dataset('tips')
plt.style.use('classic')
sb.lmplot(x='total_bill',y='tip',data=tips,fit_reg=True,hue='smoker')   #the line fit_reg
#hue to differentiate the given conditions
```

<font color='red'>col - The 'col' argument will do the same as hue but the major difference is col will plot a separate plot based on the parameters passed.
ex:- if we provide col - 'smoker' it will plot 2 plots with one plot consists of smokers and other of non smokers 

##### plotting a graph to differentiate between smokers and non smokers using col.

```python
import matplotlib.pyplot as plt
import seaborn as sb
tips=sb.load_dataset('tips')
plt.style.use('classic')
sb.lmplot(x='total_bill',y='tip',data=tips,fit_reg=True,col='smoker')
#col to get mult cols
```

### swarm plot

similar to box plot but the points in the box are visible

```python
# boxenplot

import seaborn as sb
tips=sb.load_dataset('tips')
ax =sb.boxenplot(x='day',y='total_bill',data=tips)
```

in swarm plot the box distribution is replaced by points dist. the points are swarmed around the box

##### swarm plot of tips dataset with days as x axis and total_bill as y axis

```python
# day vs total bill
import matplotlib.pyplot as plt
import seaborn as sb
tips=sb.load_dataset('tips')
plt.style.use('dark_background')
ax =sb.swarmplot(x='day',y='total_bill',data=tips)
```

### Heatmap

##### heat map using passengers as bar from flghts dataset

```python
import seaborn as sb
flights = sb.load_dataset('flights')
flights.head(1)
```

```python
flights.columns
```

```python
flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
#creating a pivot table

flights1=flights.pivot('month','year','passengers')  #passengers = bar 
flights1
```

```python
# heatmap

ax=sb.heatmap(flights1)
```

##### heatmap with annot and format

```python
import seaborn as sb
flights = sb.load_dataset('flights')
flights1=flights.pivot('month','year','passengers')  #passengers = bar 
b = sb.heatmap(flights1,annot=True,fmt='d')
```

##### Heatmap with linewidth

```python
import seaborn as sb
flights = sb.load_dataset('flights')
flights1=flights.pivot('month','year','passengers')  #passengers = bar 
b = sb.heatmap(flights1,lw=0.5)
```

##### color map-cmap on heatmap

```python
import seaborn as sb
sb.set_theme()
flights = sb.load_dataset('flights')
flights1=flights.pivot('month','year','passengers')  #passengers = bar 
b = sb.heatmap(flights1,cmap='jet',annot=True,fmt='d')
```

##### removing c bar from heatmap

```python
import seaborn as sb
flights = sb.load_dataset('flights')
flights1=flights.pivot('month','year','passengers')  #passengers = bar 
b = sb.heatmap(flights1,cbar=False)
```
